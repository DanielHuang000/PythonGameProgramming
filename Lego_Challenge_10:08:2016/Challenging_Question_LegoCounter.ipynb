```python
"""This program read a list of legos, and determine which one is bricks 
and which one is beam. beams = (3,5); bricks = (4,7)
"""

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  

def distance(a,b):
    d=(a**2+b**2)**1/2
    return d

Bricks=0
Beams=0

legos = np.loadtxt('Legos.txt',delimiter=',')

#not sure how to use these...

len(legos)
np.shape(legos)
for x, y in legos:
    plt.scatter(x,y)
    x1=abs(x-3)
    y1=abs(y-5)
    x2=abs(x-7)
    y2=abs(y-4)
    distance(x,y)
    if (distance(x1,y1)<=distance(x2,y2)):
        Beams+=1
        plt.scatter(x,y,color = 'blue')
    else:
        Bricks+=1
        plt.scatter(x,y,color = 'red')



print("Beams = " + str(Beams))
print("Bricks = " + str(Bricks))




# use to choose color
#fig, ax = plt.subplots()
#ax.set_color_cycle(['red', 'black', 'yellow'])

# use to set color
#plot(x, y, color=color)
```
