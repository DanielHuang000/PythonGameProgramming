```python


import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  



legos = np.loadtxt('Legos2.txt',delimiter=',')

Bricks=0
Beams=0

for x, y in legos:
    Dx=abs(x-0)
    Dy=abs(y-0)
    if Dx<0.5 and Dy<0.5:
        Beams+=1
        plt.scatter(x,y,color = 'blue')
    else:
        Bricks+=1
        plt.scatter(x,y,color = 'red')








print("Beams = " + str(Beams))
print("Bricks = " + str(Bricks))

```
